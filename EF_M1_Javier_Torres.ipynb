```python
'''1.	Diseñe e implemente una clase Producto con los siguientes atributos:
id, nombre, descripción, cantidad y precio.
La clase debe incluir métodos para obtener y modificar la información de los productos.'''

class Producto: #Una clase puede abarcar muchas funciones
    def __init__(self, id, nombre, descripcion, cantidad, precio):   #Atributos de instancia
        self.id = id
        self.nombre = nombre
        self.descripcion = descripcion
        self.cantidad = cantidad
        self.precio = precio

    def __str__(self):  #método que devuelve cadena que devuelve descripción
        return f"{self.nombre}, ID: {self.id}: {self.descripcion}, Cantidad: {self.cantidad}, Precio: {self.precio}" #Genera un print con f""


    def actualizar_cantidad(self, nueva_cantidad):  #Metodo que permite actualizar la cantidad del producto
        self.cantidad = nueva_cantidad

    def actualizar_precio(self, nuevo_precio):      #Metodo que permite actualizar el precio del producto
        self.precio = nuevo_precio

    def ingreso_de_producto(self, cantidad_ingresada):      #Metodo que ingresar una cantidad definida de producto
        self.cantidad += cantidad_ingresada

    def salida_de_producto(self, cantidad_salida):      #Metodo que saca una cantidad definida de producto
        self.cantidad -= cantidad_salida

    def obtener_info(self): #motodo get en ingles
        return f"ID: {self.id}, Nombre: {self.nombre}, Descripción: {self.descripcion}, Cantidad: {self.cantidad}, Precio: {self.precio}"  #Igual que _str_, solo porque lo pide el pdf


'''2.	Diseñe e implemente una clase Inventario que utilice una estructura de datos
(por ejemplo, un diccionario) para almacenar objetos de la clase Producto.
La clase Inventario debe incluir métodos para realizar las siguientes operaciones:

•	Agregar un producto al inventario
•	Eliminar un producto del inventario
•	Actualizar la información de un producto en el inventario
•	Buscar un producto en el inventario por su id
•	Listar todos los productos en el inventario, incluyendo información como la cantidad total de productos y el valor total del inventario.

'''

class Inventario:
    def __init__(self):         #Atributo de instancia o del objeto
        self.productos = {}     #Genera un diccionario con los productos

    def agregar_producto(self, producto):           #Método que permite agregar un producto ID al diccionario productos, unico
        if producto.id not in self.productos:       #Verifica si producto está en el diccionario, si no es asi, entonces:
            self.productos[producto.id] = producto  #Agrega el producto al diccionario
            return True                             #Si se realiza el agregado de producto da una respuesta True
        else:
            return False                            #Si no agrega el producto porque ya existe, devuelve False y no se agrega

    def eliminar_producto(self, id):                #Metodo que elimina un producto del diccionario
        if id in self.productos:                    #Si id existe en el diccionario productos, entra
            del self.productos[id]                  #Se elimina el producto del diccionario productos
            return True                             #Si entra al condicional entonces significa que eliminó y devuelve True
        else:
            return False                            #No existe la id de producto en productos, por lo que no se realiza la acción y devuelve False


    def actualizar_producto_cantidad(self, id, cantidad):    #Metodo que actualiza la cantidad de producto en diccionario productos, es directo
        if id in self.productos:                                  #Si id producto existe en diccionario productos, se valida
            self.productos[id].actualizar_cantidad(cantidad)      #Se actualiza ocupando el metodo de la clase producto(actualizar_cantidad)
            return True                                           #Devuelve True si entra al condicional
        else:                                                     #Devuelve False si no existe la Id del producto en el diccionario productos
            return False


    def actualizar_producto_precio(self, id, precio):        #Metodo que actualiza el precio de producto en diccionario productos, es directo
        if id in self.productos:                                  #Si id producto existe en diccionario productos, se valida
            self.productos[id].actualizar_precio(precio)      #Se actualiza ocupando el metodo de la clase producto(actualizar_precio)
            return True                                       #Devuelve True si entra al condicional
        else:
            return False                                            #Devuelve False si no existe la Id del producto en el diccionario productos

    def ingresar_cantidad_producto(self, id, cantidad_ingresada): #Metodo que permite aumentar la cantidad de producto que hay en stock
        if id in self.productos:                                  #Si id producto existe en diccionario productos, se valida
            self.productos[id].ingreso_de_producto(cantidad_ingresada) #Se agrega cantidad ocupando el metodo de la clase producto(ingreso_de_producto)
            return True                                                #Devuelve True si entra al condicional
        else:
            print("No existe ese id de producto")
            return False                                               #Devuelve False si no existe la Id del producto en el diccionario productos

    def sacar_cantidad_producto(self, id, cantidad_salida):        #Metodo que permite disminuir la cantidad de producto que hay en stock
        if id in self.productos and cantidad_salida <= self.productos[id].cantidad: #Si id producto existe en diccionario productos y la cantidad que se quiere saca no supera lo que hay en stock, se valdia
            self.productos[id].salida_de_producto(cantidad_salida)                  #Se saca cantidad ocupando el metodo de la clase producto(salida_de_producto)
            return True
        elif id in self.productos and cantidad_salida > self.productos[id].cantidad :  #Posibilidad que exista la id pero que la cantidad que se quiere sacar supere el stock actual
            print("No existe suficiente stock de este producto para el requerimiento")
            return False
        else:
            print("No existe ese id de producto")
            return False                                               #Devuelve False si no existe la Id del producto en el diccionario productos

    def buscar_producto(self, id):
        return self.productos.get(id, None)                   #https://barcelonageeks.com/metodo-get-para-diccionarios-en-python/
#        return self.productos[id]

    def listar_productos(self):
        total_de_productos = len(self.productos)                #Indica el largo de la lista, en un total de productos
        valor_total_de_productos = 0
        for producto in self.productos.values():
            valor_total_de_productos += producto.precio * producto.cantidad #Genera un valor resumido de cuanta plata hay en productos

        informe = f"Total de Productos: {total_de_productos}, Valor Total del Inventario: {valor_total_de_productos}\n"  #Un string con formato {}
        informe += "\n".join(str(producto) for producto in self.productos.values())    # join método de cadenas de texto que concatena una lista de cadenas, esto generara una lista con los productos y sus valores del diccionario
        #concatena el informe creado anteriormente, saltandose una linea y agregando la información de los productos, los \n hacen que se salte lineas
        #imprimir diccionario python
        #https://es.stackoverflow.com/questions/459471/python-como-imprimo-contenido-de-diccionario-de-forma-ordenada
        return informe


producto_1 = Producto("1", "Morocha", "Galleta Bañada Morocha 180G, 180 g'", 100, 1000)
producto_2 = Producto("2", "Triton chocolate", "Galletas Triton Chocolate, 126 g", 150, 770)
producto_3 = Producto("3", "Dos en Uno", "Galletas Bon O Bon Bolsa, 95 g", 140, 1000)
producto_4 = Producto("4", "Tuareg", "Galleta Tuareg Coco, 120 g", 100, 570)
producto_5 = Producto("5", "Mckay Vino", "Galletas Vino, 155 g", 100, 570)
producto_6 = Producto("6", "Obsesión", "Galleta Obsesión Clásica, 85 g", 100, 1090)
producto_7 = Producto("7", "Costa Donuts", "Galleta Donuts Leche, 100 g", 120, 1090)
producto_8 = Producto("8", "Chokita", "Galletas de Chocolate Bolsa, 10 Un x 30 g", 120, 3290)

inventario = Inventario()
inventario.agregar_producto(producto_1)
inventario.agregar_producto(producto_2)
inventario.agregar_producto(producto_3)
inventario.agregar_producto(producto_4)
inventario.agregar_producto(producto_5)
inventario.agregar_producto(producto_7)
inventario.agregar_producto(producto_8)

#print(inventario.listar_productos())

'''3.-Implemente una menú de usuario que permita a los usuarios interactuar con el sistema de gestión de inventario.
 El menú debe incluir opciones para realizar las diferentes operaciones disponibles en la clase Inventario.
Utilice un bucle while y estructuras de control de flujo if-elif-else para manejar las diferentes opciones del menú. (opcional)

4.-Asegúrese de que el sistema maneje adecuadamente errores y excepciones,
como entradas inválidas del usuario o intentos de actualizar o eliminar productos
que no existen en el inventario.

5.-Documente el código utilizando comentarios y docstrings apropiados para
facilitar la comprensión y el mantenimiento del software.

6.-Gestione el código a través de GitHub.
'''


def menu():                                       #Imprime un menu
    print("\n")
    print("Opciones de usuario:")
    print("1. Agregar producto")
    print("2. Eliminar producto")
    print("3. Actualizar cantidad de producto")
    print("4. Agregar cantidad de producto a stock")
    print("5. Sacar cantidad de producto de stock")
    print("6. Actualizar precio de producto")
    print("7. Buscar producto")
    print("8. Lista de todos los productos")
    print("9. Exit")


def datos_producto_agregar():      #Para ingresar productos se necesita la data y la id que debe tener, y los demás atributos
    id = input("Ingrese el ID del producto: ")
    nombre = input("Ingrese el nombre del producto: ")
    descripcion = input("Ingrese la descripción del producto: ")
    cantidad = int(input("Ingrese la cantidad del producto: "))
    precio = float(input("Ingrese el precio del producto: "))
    return Producto(id, nombre, descripcion, cantidad, precio)


while True:
    menu()
    opcion = input("Elige una opción: ")
    print("\r")

    try:
        if opcion == '1':
            producto = datos_producto_agregar()    #Para ingresar productos se necesita la data y la id que debe tener
            if inventario.agregar_producto(producto):
                print("Producto agregado")
            else:
                print("Id ya tiene un producto asignado") # Cuando usuario intenta ingresar un producto con un id ya utilizado.

        elif opcion == '2':                         #Opción para eliminar producto de inventario
            id = input("Ingrese el ID de producto para eliminar: ")
            if inventario.eliminar_producto(id):
                print("Producto eliminado")
            else:
                print("Producto no esta en el inventario")

        elif opcion == '3':                           #Opción para actualizar cantidad de producto del inventario

            id = input("Ingrese el ID del producto para actualizar cantidad: ")
            cantidad = int(input("Ingrese la nueva cantidad: "))

            if inventario.actualizar_producto_cantidad(id, cantidad):
               print("Cantidad de producto se actualizo")
            else:
               print("Producto no esta en el inventario")


        elif opcion == '4':                           #Opción para agregar producto a stock

            id = input("Ingrese el ID del producto para actualizar cantidad: ")
            cantidad_ingresada = int(input("Ingrese la cantidad que esta ingresando: "))

            if inventario.ingresar_cantidad_producto(id, cantidad_ingresada):
               print("Cantidad de producto ingresada a stock")
            else:
               print("Producto no esta en el inventario")


        elif opcion == '5':                           #Opción para sacar producto a stock

            id = input("Ingrese el ID del producto para actualizar cantidad: ")
            cantidad_salida = int(input("Ingrese la cantidad que esta sacando: "))

            if inventario.sacar_cantidad_producto(id, cantidad_salida):
               print("Cantidad de producto sacada de stock")
            else:
               print("Lo lamento")




        elif opcion == '6':                          #Opción para actualizar precio de producto del inventario

            id = input("Ingrese el ID del producto a actualizar precio: ")
            precio = float(input("Ingrese el nuevo precio: "))

            if inventario.actualizar_producto_precio(id, precio):
                print("Precio de producto se actualizo")
            else:
                print("Producto no esta en el inventario")


        elif opcion == '7':                           #Opción para buscar especificamente un producto del inventario

            id = input("Ingrese ID de producto a buscar: ")
            producto = inventario.buscar_producto(id)
            if producto:
                print(producto.obtener_info())
            else:
                print("Producto no esta en el inventario")

        elif opcion == '8':                           #Opción para generar lista de productos del inventario, la cantidad y la suma total en dinero del inventario
            print(inventario.listar_productos())

        elif opcion == '9':                           #Opcion de salida del While y del sistema
            print("Sistema se cerró")
            break

        else:                                          #Si es que pusieron una opcion que no existe
            print("No existe esa opción")

    except:                                            #Es general pero si algo falla del código, emite este mensaje
          print(f"Error de sistema")

#Producto("8", "Chokita", "Galletas de Chocolate Bolsa, 10 Un x 30 g", 120, 3290)

```
